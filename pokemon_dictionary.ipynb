```
import requests
from bs4 import BeautifulSoup
for m in range(0, 8):
    for l in range(0, 10):
        for k in range(0, 10):
          if m == 0 and l == 0 and k == 0:
            pass
          else:
            r = requests.get("https://www.pokemon.jp/zukan/detail/" + str(m) + str(l) + str(k) +  ".html")
            
            soup = BeautifulSoup(r.content, "html.parser")
            names = soup.select("p.name")
            name = names[0]
            out = name.getText()
            print(out)
            tabs = soup.select("#tab2")
            #out = tabs[0]
            for tab in tabs:
                out = tab.getText()
                print(out)
```
